```python
import pymysql
import pandas as pd

# Connect to the UCSC genome MySQL database
ucsc_db = pymysql.connect(
    host="genome-mysql.cse.ucsc.edu",
    user="genome"
)

# Run the query
result = pd.read_sql("SHOW DATABASES;", ucsc_db)

# Disconnect
ucsc_db.close()

print(result)
```

```python
import pymysql
import pandas as pd

# Connect to the hg19 database on the UCSC genome MySQL server
hg19 = pymysql.connect(
    host="genome-mysql.cse.ucsc.edu",
    user="genome",
    database="hg19"
)

# Select a specific subset
affy_mis = pd.read_sql(
    "SELECT * FROM affyU133Plus2 WHERE misMatches BETWEEN 1 AND 3",
    hg19
)

hg19.close()

# Compute quantiles of the misMatches column
print(affy_mis["misMatches"].quantile([0, 0.25, 0.5, 0.75, 1.0]))
```
